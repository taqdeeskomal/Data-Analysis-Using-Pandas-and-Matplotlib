```python
import pandas as pd
```

```python
df=pd.read_excel("Sample - Superstore (1).xls", sheet_name="Orders")
df
```

```python
df1=pd.read_excel("Sample - Superstore (1).xls", sheet_name="Returns")
df1
```

```python
df2=pd.read_excel("Sample - Superstore (1).xls", sheet_name="People")
df2
```

```python
df3=pd.concat([df,df1,df2])
df3
```

# CLEANING

```python
#drop duplicate rows
df4=df3.drop_duplicates()
df4
```

```python
#drop duplicate columns
df5=df4.T.drop_duplicates().T
df5
```

```python
#Drop NAN rows and replace NAN by zero
df6=df5.dropna(axis=0,how="all")
df6
df6.fillna(0)
```

```python
#drop NAN columns and replace NAN by zero
df7=df6.dropna(axis=1,how="all")
df7
df7.fillna(0)
```

```python
#to change columns into rows
df8=df7.T
df8
```

```python
import matplotlib.pyplot as plt
import numpy as np
#sorting by Comparison
df12=pd.DataFrame(df.groupby(['City'])['Profit'].sum().reset_index())
df12

```

```python
print("top 5 profitable cities:")
s=df12.head(5)
s
x=s["City"]
y=s["Profit"]
plt.plot(x,y)
plt.show()
```

```python
df15=pd.DataFrame(df.groupby(["Region"])["Profit"].sum().reset_index())
df15
x=df15["Region"]
y=df15["Profit"]
plt.scatter(x,y)
plt.show()
```

```python
df15=pd.DataFrame(df.groupby(["Segment"])["Profit"].sum().reset_index())
df15
x=df15["Segment"]
y=df15["Profit"]
plt.bar(x,y)
plt.show()
```

```python
category=pd.DataFrame(df.groupby(["Category"])["Profit"].sum().reset_index())
category
data=category["Profit"]
label=category["Category"]
plt.pie(data,labels=label)
plt.show()
```

```python
category=pd.DataFrame(df.groupby(['Category'])[['Sales','Profit','Quantity','Discount']].sum())
category
category.plot()
plt.show()
```
